# Create a Data Frame

```python
def com_details(soup):
    company_name= [c.text.strip() for c in soup.find_all('td',class_='companyName')]
    cin= [cin.text.strip() for cin in soup.find_all('td',class_="corporateIdentificationNumber")]
    roc=[r.text.strip() for r in soup.find_all('td',class_="roc")]
    status_link=[sl.text.strip() for sl in soup.find_all('td',class_="statusLink")]
    return pd.DataFrame({'Company Name':company_name,'CIN':cin,'ROC name':roc,'Company Status':status_link})
```

```python
dataframe=com_details(soup)
```
